```python
#importing of necessary libraries
import pandas as pd  
import numpy as np  
import seaborn as sns 
import matplotlib.pyplot as plt

import psycopg2

import warnings

warnings.filterwarnings('ignore')
```

```python
# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(
    host="localhost",  
    database="telecom",  
    user="postgres",  
    password= " " 
)

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# Write an SQL query
query = "SELECT * FROM xdr_data"

# Execute the query and fetch the data into a Pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the cursor and the connection
cur.close()
conn.close()

# Display the loaded data
df
```

```python
# Drop rows where column 'AIMEI' has NaN values
df = df.dropna(subset=['IMEI', 'IMSI'])
df
```

```python
# Calculate the threshold: At least 75% of columns (i.e., 0.75 * total columns) must be non-null
threshold = int(0.75 * df.shape[1])  # 75% threshold across columns (i.e., 1 column must be non-null)

# Drop rows where less than the threshold number of non-null values are present
df = df.dropna(thresh=threshold)

df
```

```python
# Convert 'Avg Bearer TP DL (kbps)' and 'Avg Bearer TP UL (kbps)' to Bytes per second
df['Avg Bearer TP DL (Bytes/s)'] = df['Avg Bearer TP DL (kbps)'] * 125
df['Avg Bearer TP UL (Bytes/s)'] = df['Avg Bearer TP UL (kbps)'] * 125

# Convert 'Dur. (ms)' to seconds
df['Dur. (s)'] = df['Dur. (ms)'] / 1000

# Convert 'Activity Duration DL (ms)' and 'Activity Duration UL (ms)' to seconds
df['Activity Duration DL (s)'] = df['Activity Duration DL (ms)'] / 1000
df['Activity Duration UL (s)'] = df['Activity Duration UL (ms)'] / 1000

# Now drop or keep the original columns as needed
# If you want to drop the original columns after conversion:
df = df.drop(columns=['Avg Bearer TP DL (kbps)', 'Avg Bearer TP UL (kbps)', 
                      'Dur. (ms)', 'Activity Duration DL (ms)', 'Activity Duration UL (ms)'])
```

```python
# Impute missing values
for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':  # Numerical columns
        mean_value = df[column].mean()  # Calculate median
        df[column].fillna(mean_value, inplace=True)  # Fill NaN with median
    else:  # Categorical columns
        mode_value = df[column].mode()[0]  # Calculate mode (most frequent value)
        df[column].fillna(mode_value, inplace=True)  # Fill NaN with mode

df
```

2. User Overview Analysis for Applications (xDR Sessions)

```python
# List of application columns
applications = ['Social Media', 'Google', 'Email', 'Youtube', 'Netflix', 'Gaming']

# Define aggregation functions
aggregation_functions = {
    'Bearer Id': 'count',          # Number of xDR sessions
    'Dur. (s)': 'sum',            # Session duration
    'Total DL (Bytes)': 'sum',     # Total download data
    'Total UL (Bytes)': 'sum',     # Total upload data
}

# Add application data to the aggregation
for app in applications:
    dl_col = f'{app} DL (Bytes)'
    ul_col = f'{app} UL (Bytes)'
    aggregation_functions[dl_col] = 'sum'
    aggregation_functions[ul_col] = 'sum'

# Perform aggregation per user
user_aggregate = df.groupby('MSISDN/Number').agg(aggregation_functions)
```

```python
# Rename columns for clarity
user_aggregate = user_aggregate.rename(columns={
    'Bearer Id': 'number_of_xDR_sessions',
    'Dur. (s)': 'session_duration',
    'Total DL (Bytes)': 'total_download',
    'Total UL (Bytes)': 'total_upload',
})
```

Calculate Total Data Volume per Application

```python
for app in applications:
    dl_col = f'{app} DL (Bytes)'
    ul_col = f'{app} UL (Bytes)'
    total_col = f'{app} Total Bytes'
    user_aggregate[total_col] = user_aggregate[dl_col] + user_aggregate[ul_col]
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64'])

# Compute quantiles
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = (numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))
print(outliers.sum())

# Remove outliers
df_no_outliers = df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_no_outliers.shape
```

Normalize Metrics & Run K-Means Clustering

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalized_metrics = pd.DataFrame(
    scaler.fit_transform(user_aggregate),
    columns=user_aggregate.columns,
    index=user_aggregate.index
)
```

apply K-Means clustering (with k=3):

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(normalized_metrics)
user_aggregate['Cluster'] = clusters
```

Compute Statistics for Each Cluster
To compute the minimum, maximum, average, and total for each cluster:

```python
cluster_stats = user_aggregate.groupby('Cluster').agg(
    min_duration=('session_duration', 'min'),
    max_duration=('session_duration', 'max'),
    avg_duration=('session_duration', 'mean'),
    total_duration=('session_duration', 'sum'),
    
    min_download =('total_download', 'min'),
    max_download=('total_download', 'max'),
    avg_download=('total_download', 'mean'),
    total_download=('total_download', 'sum'),

    min_upload =('total_upload', 'min'),
    max_upload=('total_upload', 'max'),
    avg_upload=('total_upload', 'mean'),
    total_upload=('total_upload', 'sum')

)
cluster_stats
```

Optimize K using the Elbow Method

```python
from sklearn.cluster import KMeans

# Determine the best value of k
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_metrics)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal k')
plt.show()
```
