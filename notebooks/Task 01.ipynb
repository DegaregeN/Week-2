```python
# Drop rows where column 'AIMEI' has NaN values
df = df.dropna(subset=['IMEI', 'IMSI'])
df
```

```python
# Calculate the threshold: At least 25% of columns (i.e., 0.25 * total columns) must be non-null
threshold = int(0.25 * df.shape[1])  # 25% threshold across columns (i.e., 1 column must be non-null)

# Drop rows where less than the threshold number of non-null values are present
df_cleaned = df.dropna(thresh=threshold)

df_cleaned
```

```python
# Impute missing values
for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':  # Numerical columns
        median_value = df[column].median()  # Calculate median
        df[column].fillna(median_value, inplace=True)  # Fill NaN with median
    else:  # Categorical columns
        mode_value = df[column].mode()[0]  # Calculate mode (most frequent value)
        df[column].fillna(mode_value, inplace=True)  # Fill NaN with mode

df
```

2. User Overview Analysis for Applications (xDR Sessions)

```python
# Aggregate per user (using MSISDN/Number as User ID)
user_aggregate = df.groupby('MSISDN/Number').agg({
    'Dur. (ms)': 'sum',
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
}).rename(columns={'Dur. (ms)': 'Session_duration', 'Total DL (Bytes)': 'Download_data', 'Total UL (Bytes)': 'Upload_data'})

# Add a column for total data (Download + Upload)
user_aggregate['Total_data'] = user_aggregate['Download_data'] + user_aggregate['Upload_data']
user_aggregate.head()
```

3. Handling Missing Values & Outliers

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64'])

# Compute quantiles
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = (numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))
print(outliers.sum())

# Remove outliers
df_no_outliers = df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_no_outliers.shape
```

4. Variable Transformations: Segment Users into Deciles

```python
# Segment users into decile classes based on total session duration
user_aggregate['Decile'] = pd.qcut(user_aggregate['Session_duration'], 10, labels=False)

# Compute total data (Download + Upload) per decile class
decile_data = user_aggregate.groupby('Decile')['Total_data'].sum()
print("Total data per decile class:\n", decile_data)
```

```python
# Plot decile data
decile_data.plot(kind='bar', title='Total Data per Decile')
plt.xlabel('Decile Class')
plt.ylabel('Total Data (Bytes)')
plt.show()
```

Bivariate Analysis

```python
# Define applications and number of subplots needed
applications = ['Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)', 
                'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  # Adjust the figsize as needed
axes = axes.flatten()  # Flatten the array of axes

# Plot each application data in a subplot
for i, app in enumerate(applications):
    axes[i].scatter(df[app], df['Total DL (Bytes)'] + df['Total UL (Bytes)'])
    axes[i].set_title(f'{app} vs Total Data')
    axes[i].set_xlabel(app)
    axes[i].set_ylabel('Total Data (Bytes)')

# Adjust layout
plt.tight_layout()
plt.show()
```

6. Correlation & Dimensionality Reduction
        
        Correlation Analysis

```python
# Compute correlation matrix for applications data
app_data = df[['Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)', 
               'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 
               'Total DL (Bytes)', 'Total UL (Bytes)']]
corr_matrix = app_data.corr()
corr_matrix

# Plot correlation matrix
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(app_data)

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Create DataFrame with principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
print(pca_df.head())

# Plot PCA results
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.title('PCA of Application Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
